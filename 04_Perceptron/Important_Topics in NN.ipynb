### Topic 00 - Different Types of Neural networks

##### i. Feed Forward neural networks -  

These are the simplest type of artificial neural network, where connections between the nodes do not form a cycle. They are primarily used for supervised learning tasks.

##### ii. Convolutional Neural Networks - 
CNNs are specialized for processing data with a grid-like topology, such as images. They use convolutional layers to automatically detect and learn spatial hierarchies of features from input images.

##### iii. Recurrent Neural networks - 

RNNs are designed to handle sequential data by having connections that form directed cycles. This makes them particularly effective for tasks like time-series prediction and natural language processing

##### iv. Long Short Term memory - 

LSTMs are a special type of RNN capable of learning long-term dependencies. They are designed to avoid the long-term dependency problem by incorporating a cell state and three gates (input, output, and forget).

##### v. Generative Adversarial networks - 

GANs consist of two neural networks: a generator and a discriminator. The generator creates data, and the discriminator evaluates it. This setup is used for generating new data samples, such as images, that resemble the training data.

##### vi. AutoEncoders - 

Autoencoders are used for unsupervised learning and data compression. They consist of an encoder that compresses the input into a lower-dimensional representation and a decoder that reconstructs the input from this representation

##### vii. Transformer Networks - 

Transformers are designed to handle sequential data and are widely used in natural language processing tasks. They rely on self-attention mechanisms to capture dependencies between different parts of the input sequence

##### viii. Radial Basis Function networks (RBF)

RBF networks use radial basis functions as activation functions and are typically used for classification and regression tasks. They are particularly effective for interpolation in high-dimensional space.

### Topic 1 - Activation Functions

Activation functions are mathematical functions used in neural networks to introduce non-linearity into the model. They determine the output of a neural network layer and help the network learn complex patterns


|Activation Function|	Description	|Formula	|When to Use|
|----------|------------|-------------|---------------|
|Sigmoid|	S-shaped curve that maps input values between 0 and 1.|	$$\sigma(x) = \frac{1}{1 + e^{-x}}$$|	Useful for binary classification problems. However, it can suffer from vanishing gradient problems.|
|Tanh	|Similar to Sigmoid but maps input values between -1 and 1.	|$$\tanh(x) = \frac{e^x - e{-x}}{ex + e^{-x}}$$	|Preferred over Sigmoid in practice as it centers data around zero and has stronger gradients.|
|ReLU (Rectified Linear Unit)	|ReLU outputs the input directly if positive, otherwise, it outputs zero.	|$$\text{ReLU}(x) = \max(0, x)$$	|Commonly used in hidden layers of neural networks. Efficient and reduces the likelihood of vanishing gradient problems.|
|Leaky ReLU|	A variation of ReLU that allows a small, non-zero gradient when the input is negative.	|$$\text{Leaky ReLU}(x) = \max(0.01x, x)$$	|Used to address the "dying ReLU" problem where neurons can sometimes get stuck during training.|
|Parametric ReLU (PReLU)	|A generalization of Leaky ReLU where the slope of the negative part is learned during training.|	$$\text{PReLU}(x) = \max(\alpha x, x)$$, where $$\alpha$$ is a learned parameter.|	Provides greater flexibility and may lead to better performance compared to ReLU and Leaky ReLU.|
|ELU (Exponential Linear Unit)	|Similar to ReLU but uses an exponential function for negative inputs.	|$$\text{ELU}(x) = \begin{cases} x & \text{if } x > 0 \\ \alpha(e^x - 1) & \text{if } x \le 0 \end{cases}$$	|Helps with faster and more accurate learning by avoiding the issue of dying ReLUs.|
|Softmax	|Generalizes the Sigmoid function for multi-class classification. It outputs a probability distribution.|	$$\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$$	|Used in the output layer of a neural network for multi-class classification problems.|

- Sigmoid and Tanh are good for binary classification but can suffer from vanishing gradients.
- ReLU is widely used in hidden layers due to its efficiency but can cause "dying ReLUs."
- Leaky ReLU and PReLU address the dying ReLU problem and offer more flexibility.
- ELU offers better learning dynamics compared to ReLU.
- Softmax is specifically for multi-class classification tasks.

### Topic 4 - batch size

Batch size in neural networks refers to the number of training examples used in one iteration to update the model parameters. It plays a critical role in the training process, impacting convergence speed, model performance, and computational efficiency

- Medium Batch Size: Typically ranging from 32 to 256.
- Large Batch Size: Typically greater than 256.

### Topic 5 - Learning rate and Gradient Descent

The learning rate in a neural network is a hyperparameter that controls how much the model's weights are adjusted with respect to the gradient of the loss function. It determines the step size taken during the optimization process to minimize the loss.

The ideal learning rate depends on the specific problem, dataset, and model architecture. It often requires experimentation to find the best value. Common practices include:

- Small Learning Rate: Often in the range of 1e-4 to 1e-2.
- Moderate Learning Rate: Typically around 1e-2 to 1e-1.
- High Learning Rate: Greater than 1e-1.


|Learning Rate	|Advantages	|Disadvantages	|When to Use|
|-------------|---------------|---------------|---------------|
|Small Learning Rate (e.g., 1e-4)	|Ensures gradual and stable convergence. |Less likely to undershoot the minimum.	|Slow convergence, may require more epochs to train.	|When dealing with complex models or datasets, to ensure stability.|
|Moderate Learning Rate (e.g., 1e-2)	|Balances between convergence speed and stability.	|May still be slow for very large datasets, requires tuning.	|When seeking a balanced approach for general tasks.|
|High Learning Rate (e.g., 1e-1 or higher)	|Speeds up convergence, quickly reaches a minimum.	|Risk of overshooting the minimum, can cause instability in training.	|When working with simple models or seeking faster results, provided stability is managed.|

### Topic 6 - warm start

A warm start in neural networks refers to the practice of initializing the training of a new model using the parameters (weights and biases) from a previously trained model, rather than starting from scratch. This approach is often used when new data becomes available and the model needs to be updated

##### Why Warm start is needed?
- Faster Convergence: By starting with pre-trained parameters, the model can converge faster because it doesn't have to learn from random initial weights. This can save computational resources and time.

- Better Generalization: Warm starting can help the model generalize better to new data, especially when the new data is similar to the old data. This can improve the model's performance on unseen data.

- Handling Incremental Data: In scenarios where data arrives incrementally (e.g., daily financial data), warm starting allows the model to continuously update without starting from scratch each time.

Sometimes, starting from scratch (cold start) might yield better results, especially if the new data is significantly different from the old data. Researchers have proposed techniques like the "shrink and perturb trick" to overcome some of the limitations of warm starting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Generate a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

momentum_values = [0.0, 0.3, 0.6, 0.9]
train_scores = []
test_scores = []

for momentum in momentum_values:
    mlp = MLPClassifier(hidden_layer_sizes=(50,), 
                        max_iter=1000, 
                        solver='sgd', 
                        momentum=momentum, 
                        random_state=42)
    
    mlp.fit(X_train, y_train)

    train_scores.append(mlp.score(X_train, y_train))
    test_scores.append(mlp.score(X_test, y_test))

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(momentum_values, train_scores, label='Training Accuracy')
plt.plot(momentum_values, test_scores, label='Testing Accuracy')
plt.xlabel('Momentum')
plt.ylabel('Accuracy')
plt.title('Effect of Momentum on MLP Performance')
plt.legend()
plt.grid(True)
plt.show()

```

### Topic 8 - Early Stopping

Early stopping is a technique used in training neural networks to prevent overfitting. Overfitting occurs when a model learns not only the underlying patterns in the training data but also the noise, leading to poor generalization on new, unseen data. Early stopping addresses this by halting the training process once the model's performance on a validation set starts to deteriorate.

##### Why Early Stopping is Required:
- Prevent Overfitting: By stopping the training process before the model overfits, it helps in achieving better generalization on new data.

- Save Computational Resources: Early stopping can reduce the training time by avoiding unnecessary epochs that do not contribute to better performance.

- Improved Model Performance: Helps in maintaining a balance between model complexity and training time, leading to more reliable models.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.callbacks import EarlyStopping

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize the data
X_train = X_train / 255.0
X_test = X_test / 255.0

def create_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Create the model
model_es = create_model()

# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model with early stopping
history_es = model_es.fit(X_train, y_train, validation_split=0.2, epochs=50, callbacks=[early_stopping], verbose=1)

# Create the model
model_no_es = create_model()

# Train the model without early stopping
history_no_es = model_no_es.fit(X_train, y_train, validation_split=0.2, epochs=50, verbose=1)

# Plot training & validation loss values
plt.figure(figsize=(12, 6))

# With early stopping
plt.subplot(1, 2, 1)
plt.plot(history_es.history['loss'], label='Training Loss')
plt.plot(history_es.history['val_loss'], label='Validation Loss')
plt.title('Model Loss with Early Stopping')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Without early stopping
plt.subplot(1, 2, 2)
plt.plot(history_no_es.history['loss'], label='Training Loss')
plt.plot(history_no_es.history['val_loss'], label='Validation Loss')
plt.title('Model Loss without Early Stopping')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

### Topic 9 - beta and epsilon

##### Beta (Œ≤)

Beta (Œ≤) is commonly used in optimization algorithms such as Momentum and Adam (Adaptive Moment Estimation).

Momentum:

In the momentum optimization algorithm, beta is the momentum coefficient that determines the contribution of the previous gradients to the current update.

Higher values of beta (e.g., 0.9 or 0.99) give more weight to the past gradients, leading to smoother and more stable updates.

Momentum equation with beta: $$v_t = \beta v_{t-1} + (1 - \beta) \nabla J(\theta_{t-1})$$

In the Adam optimizer, beta appears twice, as beta1 (Œ≤1) and beta2 (Œ≤2).

Œ≤1 controls the exponential decay rate for the first moment estimates (mean of the gradients).

Œ≤2 controls the exponential decay rate for the second moment estimates (uncentered variance of the gradients).

Adam equations with beta: $$m_t = \beta_1 m_{t-1} + (1 - \beta_1) \nabla J(\theta_{t-1})$$ $$v_t = \beta_2 v_{t-1} + (1 - \beta_2) (\nabla J(\theta_{t-1}))^2$$

### Topic 10 - vanishing gradient

The vanishing gradient problem is a significant challenge in training deep neural networks, particularly those with many layers, such as Recurrent Neural Networks (RNNs) and deep feedforward networks. It occurs when the gradients of the loss function with respect to the model parameters (weights) become exceedingly small, effectively preventing the network from learning.

In neural networks, training involves backpropagating the error (or loss) through the layers to update the weights. During backpropagation, the gradients are calculated at each layer. These gradients are then multiplied by the weights to update the weights for the next iteration.

In deep networks, as gradients are propagated back through many layers, they can become very small due to repeated multiplication by values less than one (typically, the derivatives of activation functions like sigmoid or tanh). This results in the gradients diminishing exponentially, making them too small to significantly update the weights of the initial layers. Consequently, these early layers learn very slowly or not at all.

Mathematically, if the activation function's derivative is small (e.g., sigmoid or tanh), then during backpropagation: $$\frac{\partial L}{\partial w} = \frac{\partial L}{\partial z} \cdot \frac{\partial z}{\partial w}$$

##### How to Avoid the Vanishing Gradient Problem:
1. Activation Functions
ReLU (Rectified Linear Unit): Using ReLU or its variants like Leaky ReLU helps prevent the gradient from vanishing because they don't saturate.

Swish and Mish: These newer activation functions also address the vanishing gradient problem effectively.

2. Weight Initialization
- He Initialization: Also known as Kaiming initialization, it‚Äôs suitable for networks with ReLU activations.
- Xavier Initialization: Also known as Glorot initialization, it works well with tanh and sigmoid activations.

3. Batch Normalization: By normalizing the input of each layer, batch normalization helps mitigate the vanishing gradient problem and accelerates training.

4. Residual Networks (ResNets)
Skip Connections: These connections allow the gradient to flow directly through the network, bypassing some layers, thus mitigating the vanishing gradient issue.

5. Gradient Clipping
Clip Gradients: By clipping the gradients during backpropagation, we can prevent them from becoming too small (or too large).

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# 1. Define a simple neural network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(1, 10)
        self.fc2 = nn.Linear(10, 10)
        self.fc3 = nn.Linear(10, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.sigmoid(self.fc1(x))
        x = self.sigmoid(self.fc2(x))
        x = self.sigmoid(self.fc3(x))
        return x

# 2. Initialize the network, loss function, and optimizer
net = SimpleNN()
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.1)

# 3. Generate some data
x = torch.linspace(-10, 10, 100).view(-1, 1)
y = torch.sin(x)

# 4. Track gradients during training
gradients = []

for epoch in range(100):
    optimizer.zero_grad()
    outputs = net(x)
    loss = criterion(outputs, y)
    loss.backward()

    # Save the gradient of the first layer
    gradients.append(net.fc1.weight.grad.abs().mean().item())

    optimizer.step()

# 5. Plot the gradients
plt.plot(gradients)
plt.xlabel('Epoch')
plt.ylabel('Mean Absolute Gradient')
plt.title('Vanishing Gradient Problem')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt

class CustomNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, activation_func='relu'):
        super(CustomNetwork, self).__init__()

        # Linear layers
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, output_size)

        # Activation function
        if activation_func == 'relu':
            self.activation = nn.ReLU()
        elif activation_func == 'leaky_relu':
            self.activation = nn.LeakyReLU(negative_slope=0.01)
        elif activation_func == 'prelu':
            self.activation = nn.PReLU()
        elif activation_func == 'elu':
            self.activation = nn.ELU()
        else:
            raise ValueError(f"Unsupported activation function: {activation_func}")

    def forward(self, x):
        x = self.activation(self.fc1(x))
        x = self.activation(self.fc2(x))
        x = self.fc3(x)
        return x

# Example usage with different activations
input_size = 10
hidden_size = 20
output_size = 5

# ReLU activation
model_relu = CustomNetwork(input_size, hidden_size, output_size, activation_func='relu')

# Leaky ReLU activation
model_leaky_relu = CustomNetwork(input_size, hidden_size, output_size, activation_func='leaky_relu')

# PReLU activation
model_prelu = CustomNetwork(input_size, hidden_size, output_size, activation_func='prelu')

# ELU activation
model_elu = CustomNetwork(input_size, hidden_size, output_size, activation_func='elu')

# Training loop and gradient tracking
def train_and_track_gradients(model):
    optimizer = optim.SGD(model.parameters(), lr=0.01)
    criterion = nn.MSELoss()
    
    # Dummy data
    x = torch.randn(64, input_size)
    y = torch.randn(64, output_size)
    
    gradients = []

    for epoch in range(100):
        optimizer.zero_grad()
        outputs = model(x)
        loss = criterion(outputs, y)
        loss.backward()

        # Save the gradient of the first layer
        gradients.append(model.fc1.weight.grad.abs().mean().item())
        
        optimizer.step()
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch + 1}/100], Loss: {loss.item():.4f}')
    
    return gradients

# Function to plot gradients
def plot_gradients(gradients, activation_name):
    plt.plot(gradients)
    plt.xlabel('Epoch')
    plt.ylabel('Mean Absolute Gradient')
    plt.title(f'Gradients over Epochs ({activation_name})')
    plt.show()

# Train each model and plot the gradients
print("Training with ReLU activation:")
gradients_relu = train_and_track_gradients(model_relu)
plot_gradients(gradients_relu, 'ReLU')

print("\nTraining with Leaky ReLU activation:")
gradients_leaky_relu = train_and_track_gradients(model_leaky_relu)
plot_gradients(gradients_leaky_relu, 'Leaky ReLU')

print("\nTraining with PReLU activation:")
gradients_prelu = train_and_track_gradients(model_prelu)
plot_gradients(gradients_prelu, 'PReLU')

print("\nTraining with ELU activation:")
gradients_elu = train_and_track_gradients(model_elu)
plot_gradients(gradients_elu, 'ELU')
```
